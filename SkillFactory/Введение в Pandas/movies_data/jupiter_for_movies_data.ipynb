```python
import pandas as pd
original_ratings1=pd.read_csv('ratings1.csv', sep=',')
original_ratings2=pd.read_csv('ratings2.csv', sep=',')
original_movies=pd.read_csv('movies.csv', sep=',')
original_dates=pd.read_csv('dates.csv',sep=',')

ratings1=original_ratings1.copy()
ratings2=original_ratings2.copy()
movies=original_movies.copy()
dates=original_dates.copy()

dates['date']=pd.to_datetime(dates['date'])
dates['year']=dates['date'].dt.year
dates['year'].value_counts(normalize=True)

ratings=pd.concat([ratings1, ratings2], ignore_index=True)
ratings=ratings.drop_duplicates(ignore_index=True)
ratings_dates=pd.concat([ratings,dates], axis=1)
ratings_dates.tail(7)
ratings_dates=ratings_dates.drop(['year'], axis=1)
joined=ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
merged

```
